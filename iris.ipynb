```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/home/mike/Downloads/data/archive(13)/Iris.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data["Species"] = data["Species"].astype("category")
```

```python
data.info()
```

```python
plt.figure(figsize=(15,15))

plt.subplot(221)
sns.boxenplot(data['SepalLengthCm'], y=data['Species'])

plt.subplot(222)
sns.boxenplot(data['SepalWidthCm'], y=data['Species'])

plt.subplot(223)
sns.boxenplot(data['PetalLengthCm'], y=data['Species'])

plt.subplot(224)
sns.boxenplot(data['PetalWidthCm'], y=data['Species'])
plt.tight_layout()
plt.show()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(221)
sns.swarmplot(data['SepalLengthCm'],data['SepalWidthCm'], hue=data['Species'])

plt.subplot(222)
sns.swarmplot(data['PetalLengthCm'], data['PetalWidthCm'], hue=data["Species"])
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.jointplot(data['SepalLengthCm'],data['SepalWidthCm'], kind="hex")

sns.jointplot(data['PetalLengthCm'], data['PetalWidthCm'], kind="hex")
plt.show()
```

```python
sns.kdeplot(data["PetalWidthCm"])
```

```python
plt.figure(figsize=(10,7))
sns.countplot(data=data, x='Species')
```
